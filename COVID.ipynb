```python
import sqlite3
import pandas as pd
con = sqlite3.connect('hospital.db')
cur = con.cursor()
```

```python
data_death = pd.read_csv('/content/sample_data/CovidDeaths.csv')
data_death.columns = data_death.columns.str.strip()
```

```python
data_death.to_sql("deaths", con)
```

```python
data_vac = pd.read_csv('/content/sample_data/CovidVaccinations.csv')
data_vac.columns = data_vac.columns.str.strip()
```

```python
data_vac.to_sql("vaccinations", con)
```

```python
query = '''SELECT location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 AS death_rate_in_cases
           FROM deaths
           WHERE location = "United States"
           ORDER BY 1, 2 '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT location, date, total_cases, population, ROUND((total_cases/population)*100, 2)  AS case_rate
           FROM deaths
           WHERE location = "United States"
           ORDER BY 1, 2 '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT location, MAX(total_cases) AS highest_number_of_cases,
                  MAX(ROUND((total_cases/population)*100, 2)) AS max_case_rate
           FROM deaths
           GROUP BY location
           ORDER BY max_case_rate '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT location, MAX(total_deaths) AS total_death_count
           FROM deaths
           WHERE continent IS NOT NULL
           GROUP BY location
           ORDER BY total_death_count '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT location, MAX(total_deaths) AS total_death_count
           FROM deaths
           WHERE continent IS NULL
           GROUP BY location
           ORDER BY total_death_count '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT date, SUM(new_cases) AS new_cases, SUM(new_deaths) AS new_deaths, SUM(new_deaths)/SUM(new_cases) AS death_percentage
           FROM deaths
           WHERE continent IS NOT NULL
           GROUP BY date
           ORDER BY date '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT *
           FROM deaths
           JOIN vaccinations
           ON deaths.date = vaccinations.date AND deaths.location = vaccinations.location '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT deaths.continent, deaths.location, deaths.date, deaths.population, vaccinations.new_vaccinations
           FROM deaths
           JOIN vaccinations
           ON deaths.date = vaccinations.date AND deaths.location = vaccinations.location
           WHERE deaths.continent IS NOT NULL 
           ORDER BY 1, 2 '''
data = pd.read_sql(query, con)
data
```

```python
query = '''SELECT deaths.continent, deaths.location, deaths.date, deaths.population, vaccinations.new_vaccinations,
             SUM(vaccinations.new_vaccinations) OVER (PARTITION BY deaths.location ORDER BY deaths.location, deaths.date)
             AS rolling_people_vaccinated,
             ((SUM(vaccinations.new_vaccinations) OVER (PARTITION BY deaths.location ORDER BY deaths.location, deaths.date))/population)*100 AS vaccination_rates
           FROM deaths
           JOIN vaccinations
           ON deaths.date = vaccinations.date AND deaths.location = vaccinations.location
           WHERE deaths.continent IS NOT NULL
           ORDER BY 1, 2 '''
data = pd.read_sql(query, con)
data
```
